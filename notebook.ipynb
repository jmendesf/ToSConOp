```python
import copy
from simplificationToS import *
import numpy as np
from skimage import io

# The mage MUST be a single channel greyscale image
urban = 'images/urban.png'
naia = 'images/naia.png'
aorta = 'images/aorte.png'

image_urban = io.imread(urban)
image_urban = np.pad(image_urban, 1, mode="constant", constant_values=0)
imshow(image_urban,cmap='gray', figsize=(300,300))
tos_urban = TreeOfShapes(image_urban)   
print("Urban: tree of shapes composed of", tos_urban.get_nb_nodes(), "nodes.")

image_naia = io.imread(naia)
image_naia = np.pad(image_naia, 1, mode="constant", constant_values=0)
imshow(image_naia,cmap='gray', figsize=(300,300))
tos_naia = TreeOfShapes(image_naia)   
print("Naia: tree of shapes composed of", tos_naia.get_nb_nodes(), "nodes.")
```

```python
tos_proper_part = copy.deepcopy(tos_urban) 

tos_proper_part.filter_tree_proper_part_bottom_up_consecutive(2, 1, 11)
print("nb nodes:", str(tos_proper_part.get_nb_nodes()))
    
image_proper_part = tos_proper_part.reconstruct_image()
imshow(image_proper_part,cmap='gray', figsize=(300,300))
image_proper_part.save("output/image_proper_part.png")
```

```python
tos_mean_filtered = copy.deepcopy(tos_naia)

for i in range(0,10):
    tos_mean_filtered.filter_tree_mean()

image_mean_filtered = tos_mean_filtered.reconstruct_image()
imshow(image_mean_filtered,cmap='gray', figsize=(300,300))
image_mean_filtered.save("output/image_mean_filtered.png")
```
